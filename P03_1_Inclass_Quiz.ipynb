```python
titles = pd.read_csv('data/titles.csv', index_col=None)
titles.head()
```

```python
# 인클 퀴즈 2번
# 문제 1번에서의 titles 데이터프레임을 사용하여 아래 질문에 답을 하시오.
# Hamlet이라는 타이틀을 갖고 로칭된 필름은 각 10년간 얼마나 있는가? (groupby 메소드를 사용할 것.)
titles = titles[titles.title =='Hamlet']
titles

titles['year'] = titles.year // 10 * 10
titles.groupby('year').count()
```
