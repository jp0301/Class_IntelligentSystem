```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
plt.rc('font', family='Malgun Gothic')
plt.rc("axes",unicode_minus=False)
from __future__ import division
```

```python
df = pd.read_excel("19daeseon.xlsx",encoding = "cp949")
```

```python
df.head()
```

```python
df.info()
```

```python
provincedf = df[df['구시군명'] == '합계'].drop(["읍면동명", "투표구명", "선거인수", "투표수", "계", "무효투표수", "기권수"], 1).reset_index().drop(['index','구시군명'],1).set_index('시도명')
provincedf
```

```python
provincedf.plot(kind='barh',figsize=(10,12))
```

```python
provincedf.plot.hist(bins=20, alpha=0.5)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
influential=provincedf.T.head(5).T
influential
```
